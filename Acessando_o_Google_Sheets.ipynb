Autenticação para uso do google sheets


```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

Editando primeira pagina da planilha

```python
spreadsheet  = gc.open('exames_soror_cvan_junho_2025')
```

```python
page1 = spreadsheet.sheet1
```

```python
page1.row_values(1) #page.row_values(1) imprime a primeira linha, page.col_values(1) imprime a primeira coluna
```

```python
import pandas as pd
```

```python
data = page1.get_all_values()
headers = data[4]  # Pega a primeira linha como cabeçalho
records = data[5:]  # Pega o restante como dados
exames_soror_cvan_junho_2025 = pd.DataFrame(records, columns=headers)
```

```python
exames_soror_cvan_junho_2025.head()
```

```python
print(exames_soror_cvan_junho_2025.columns.tolist())
```

```python
exames_soror_cvan_junho_2025['Tutor'] = exames_soror_cvan_junho_2025['Tutor'].apply(
    lambda x: x.split()[0] if isinstance(x, str) and x.strip() else ''
)

exames_soror_cvan_junho_2025['Paciente'] = exames_soror_cvan_junho_2025['Paciente'].apply(
    lambda x: x.split()[0] if isinstance(x, str) and x.strip() else ''
)
```

```python
exames_soror_cvan_junho_2025.head()
```

```python
# Supondo que os nomes das colunas já estejam padronizados e corretos
tabela_agrupada = exames_soror_cvan_junho_2025.groupby(['Tutor', 'Paciente'])['Exames'].agg(lambda x: ''.join(x)).reset_index()
```

```python
tabela_agrupada.sort_values(by=['Tutor'])
```

Editando a segunda pagina da planilha


```python
page2 = spreadsheet.worksheet("SimplesVet")
df = pd.DataFrame(page2.get_all_records())
df.head()
```

```python
data = page2.get_all_values()
headers = data[0]  # Pega a primeira linha como cabeçalho
records = data[1:]  # Pega o restante como dados
exames_simplesvet_cvan_junho_2025 = pd.DataFrame(records, columns=headers)
```

```python
exames_simplesvet_cvan_junho_2025.head()
```

```python
exames_simplesvet_cvan_junho_2025['Cliente'] = exames_simplesvet_cvan_junho_2025['Cliente'].apply(
    lambda x: x.split()[0] if isinstance(x, str) and x.strip() else ''
)

exames_simplesvet_cvan_junho_2025['Animal'] = exames_simplesvet_cvan_junho_2025['Animal'].apply(
    lambda x: x.split()[0] if isinstance(x, str) and x.strip() else ''
)
```

```python
tabela_agrupada_simplesvet = exames_simplesvet_cvan_junho_2025.groupby(['Cliente', 'Animal'])['Produto/serviço'].agg(lambda x: ', '.join(x)).reset_index()
```

```python
tabela_agrupada_simplesvet.sort_values(by=['Cliente'])
```
