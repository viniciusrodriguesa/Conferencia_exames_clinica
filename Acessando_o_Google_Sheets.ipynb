```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
spreasheet = gc.open('exames_soror_cvan_junho_2025')
```

```python
page = spreasheet.sheet1
```

```python
page.row_values(1) #page.row_values(1) imprime a primeira linha, page.col_values(1) imprime a primeira coluna
```

```python
import pandas as pd
```

```python
data = page.get_all_values()
headers = data[0]  # Pega a primeira linha como cabeçalho
records = data[5:]  # Pega o restante como dados
exames_soror_cvan_junho_2025 = pd.DataFrame(records, columns=headers)
```
